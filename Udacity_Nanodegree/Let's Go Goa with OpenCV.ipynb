```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = cv2.imread('group_bluescreen.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
image_copy = np.array(image)
lower_bound = np.array([0,0,100])
upper_bound = np.array([50,50,255])
mask = cv2.inRange(image_copy,lower_bound,upper_bound)
```

```python
masked_image = np.copy(image_copy)
masked_image[mask != 0] = [0,0,0]
background_img = cv2.imread('beaches.jpg')
background_img = cv2.cvtColor(background_img, cv2.COLOR_BGR2RGB)
background_img = background_img[:532,:1152]
background_copy = np.copy(background_img)
```

```python
background_copy[mask == 0] = [0,0,0]
plt.imshow(background_copy)
```

```python
complete_img = masked_image + background_copy
plt.imshow(complete_img)
```
